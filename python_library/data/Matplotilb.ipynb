```python
# 여러개의 라인 그래프 그리기 
# 그래프의 종류를 정리하는 방법 ->> 법례 legend()

x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')

plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')

# 각 축에 대한 타이틀 달아주기
plt.xlabel('x')
plt.ylabel('y', rotation = 0)

# 그래프에 대한 제목 달기
plt.title('x,y,z datas')
plt.show()
```

```python
# bar 그래프 그리기
import numpy as np
```

```python
x = np.arange(6)
y = [80, 85, 70, 60, 50, 90]

x_label = ['a','b','c','d','e','f']

plt.bar(x, y)
plt.xticks(x, x_label)

# 실제 데이터에 대한 수치값 표현하기
for i in range(6):
    plt.text(x[i],y[i],f'{y[i]}', color='red' # text(x좌표, y좌표, 텍스트)

plt.show()
```
