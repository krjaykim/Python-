```python
df2016 = pd.read_csv('data/2016.csv', encoding = 'euc-kr', index_col = '관서명')
df2017 = pd.read_csv('data/2017.csv', encoding = 'euc-kr', index_col = '관서명')
```

```python
# 2015-2017 사이의 구분이 발생건수인 내용들만 추출해 오기!
df2015[df2015['구분'] == '발생건수']
```

```python
data1 = df2015[df2015['구분'] == '발생건수']
```

```python
# 2016, 2017 발생건수 내용만 추출하기
data2 = df2016[df2016['구분'] == '발생건수']
data3 = df2017[df2017['구분'] == '발생건수']
```

```python
display(data2)
display(data3)
```

```python
data1['2015총계'] = data1.loc[:,'살인':'폭력'].sum(axis = 1 )
```

```python
data1
```

```python
data2['2016총계'] = data2.loc[:,'살인':'폭력'].sum(axis = 1 )
data3['2017총계'] = data3.loc[:,'살인':'폭력'].sum(axis = 1 )
display(data2)
display(data3)
```

```python
data1 = data1.loc[:,'2015총계']
data1
```

```python
data2 = data2.loc[:,'2016총계']
data2
```

```python
data3 = data3.loc[:,'2017총계']
data3
```

```python
# 증감률 계산하기
# (금년 - 작년)/작년 * 100

result1 = (data2-data1)/data1 * 100
result1
```

```python
# 2. 2016=2017 증감률

result2 = (data3 - data2)/data2 * 100
result2
```

```python
# 각각 Serires 존재하는 데이터 병합하기 >>> concat()
total = pd.concat([data1, result1, data2, result2, data3], axis = 1)
total
```

```python
# column 이름 수정하기 >> rename()
total.rename(columns = {0 : '2015-2016 증감률', 1: '2016-2017 증감률'}, inplace = True)
```

```python
total
```

```python
# 결축치가 존재하는 '광주지방경찰청' 행 삭제하기 >> drop()
total.drop('광주지방경찰청',inplace=True)
```
